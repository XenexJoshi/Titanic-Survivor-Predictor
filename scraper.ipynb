```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url_vic = 'https://titanicfacts.net/titanic-casualties-list/'
page_vic = requests.get(url_vic)
soup_vic = BeautifulSoup(page_vic.text, 'html')
```

```python
print(soup_vic)
```

```python
table_vic = soup_vic.find_all('table')
print(len(table_vic))
```

```python
headers_vic = [header.get_text(strip=True) for header in table_vic[0].find_all('td', class_ = 'tdheading')]
print(headers_vic)
```

```python
data_vic = []
for index, table in enumerate(table_vic):
  for row in table.find_all('tr')[1:]:
    items = [item.get_text(strip = True) for item in row.find_all('td')]
    data_vic.append(items)

print(data_vic)
```

```python
df_vic = pd.DataFrame(data_vic, columns = headers_vic)
df_vic.head()
```

```python
df_vic['temp'] = df_vic['First Names'].str.split(' ').str[0]
df_vic.head()
```

```python
def categorize_sex(salutation):
  if salutation == "Mr" or salutation == "Master":
    return "Male"
  else:
    return "Female"

df_vic['Sex'] = df_vic['temp'].apply(categorize_sex)
df_vic.drop('temp', axis = 1, inplace = True)
df_vic.head()
```

```python
df_vic['Status'] = 'Dead'
df_vic.head()
```

```python
df_vic.to_csv('data/victim.csv', index = False)
```

```python
url_sur = 'https://titanicfacts.net/titanic-survivors-list/'
page_sur = requests.get(url_sur)
soup_sur = BeautifulSoup(page_sur.text, 'html')
```

```python
print(soup_sur)
```

```python
table_sur  = soup_sur.find_all('table') 
print(len(table_sur))
```

```python
headers_sur = [header.get_text(strip = True) for header in table_sur[0].find_all('td', class_ = 'tdheading')]
print(headers_sur)
```

```python
data_sur = []
for index, table in enumerate(table_sur):
  for row in table.find_all('tr')[1:]:
    items = [item.get_text(strip = True) for item in row.find_all('td')]
    data_sur.append(items)

print(data_sur)
```

```python
df_sur = pd.DataFrame(data_sur, columns = headers_sur)
df_sur.head()
```

```python
df_sur['temp'] = df_sur['First Names'].str.split(' ').str[0]
df_sur['Sex'] = df_sur['temp'].apply(categorize_sex)
df_sur.drop('temp', axis = 1, inplace = True)
df_sur.head()
```

```python
df_sur['Status'] = 'Alive'
df_sur.head()
```

```python
df_sur.to_csv('data/survivor.csv', index = False)
```

```python
df_vic.drop('Body Number', axis = 1, inplace = True)
df_sur.drop('Lifeboat #', axis = 1, inplace = True)
```

```python
df = pd.concat([df_vic, df_sur], ignore_index = True)
df = df.sort_values(by = 'Surname')
df.head()
```

```python
df.to_csv('Titanic_passengers.csv', index = False)
```
